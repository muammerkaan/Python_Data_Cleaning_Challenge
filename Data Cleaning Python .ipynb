```python
# Import Dataset Mac
df = pd.read_csv('data/data_cleaning_challenge.csv')
```

```python
df
```

```python
# Drop empty columns
df = df.drop(["Unnamed: 9", "Unnamed: 10"], axis = 1)
df
```

```python
# Drop 'Row Type' NA values
df = df[df['Row Type'].notna()]
df
```

```python
# Assign a number to each person
column_values = []
iteration = 0

for i in df['Row Type']:
    if 'first name' in i:
        iteration += 1
    column_values.append(iteration)    
```

```python
# Add column to dataset to identify each person
df['Iteration'] = column_values
df
```

```python
# Drop extra column headers
df = df[df['Row Type'] != 'Row Type']
df
```

```python
# Exract first name, last name and date from df into a new dataframe
name_df = df[df['Row Type'].str.contains('first name')]
name_df.drop(columns = ['Speed1', 'Speed2', 'Electricity', 'Effort', 'Weight', 'Torque'], inplace = True)
name_df.rename(columns = {'Row Type': 'First Name', 'Iter Number': 'Last Name', 'Power1': 'Date'}, inplace = True)
name_df['First Name'] = name_df['First Name'].str[12:]
name_df['Last Name'] = name_df['Last Name'].str[11:]
name_df['Date'] = name_df['Date'].str[6:]
name_df
```

```python
# Drop rows with first name
df = df[~df['Row Type'].str.contains('first name')]
df
```

```python
# Merge name_df and df
df = pd.merge(left = name_df, right = df, how = 'inner', on = 'Iteration')
df
```
